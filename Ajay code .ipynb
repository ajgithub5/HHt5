```python
df = df.dropna()
```

```python
for x in list(df):
    df[x] = df[x].astype(float)
```

```python
dist = []
for x in range(df.shape[0]):
    dist.append(euclidean(test,df.iloc[x]))
df["distances"] = dist
```
